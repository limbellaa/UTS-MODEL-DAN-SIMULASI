```python
import pandas as pd
from sklearn.datasets import load_iris

from google.colab import drive

drive.mount('/content/drive')
directory= '/content/drive/MyDrive/survey lung cancer.csv'

iris = pd.read_csv(directory)
```

```python
iris.info()

iris.head()
```

```python
X = iris[['AGE','SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE','CHRONIC DISEASE','FATIGUE ','ALLERGY ','WHEEZING','ALCOHOL CONSUMING','COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']]
y = iris['LUNG_CANCER']

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
```

```python
tree_model = tree_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(X_test)

acc_score = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_score)
print('classification report')
print(classification_report(y_pred,  y_test))
```
